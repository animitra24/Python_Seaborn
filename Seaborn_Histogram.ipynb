# Seaborn Histogram

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
#load dataset using sns
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
#distplot will show data as histogram
sns.distplot(df["total_bill"])
```

```python
sns.distplot(df["total_bill"], bins=100, hist=False)
```

```python
sns.distplot(df["total_bill"], bins=100, hist=True, kde=False, rug=True)
```

```python
sns.distplot(df["total_bill"], bins=100, hist=True, kde=True, rug=True, fit=norm, color='y', vertical=True)
```

```python
sns.distplot(df["total_bill"], bins=100, hist=True, kde=True, rug=False, fit=norm, color='r', 
             vertical=False, axlabel="Total Bill", label="Resturant")
plt.title("Explore Resturant")
plt.legend()
```
