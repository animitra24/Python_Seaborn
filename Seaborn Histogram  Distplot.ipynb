# Seaborn Histogram / Distplot

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
#load dataset using sns
df = sns.load_dataset("tips")
```

```python
df.head()
```

```python
#distplot will show data as histogram
sns.distplot(df["total_bill"])
```

```python
sns.distplot(df["total_bill"], bins=100,hist=False )
```

```python
sns.distplot(df["total_bill"], bins=100, hist=True, kde=False, rug=True)
```

```python
sns.distplot(df["total_bill"], bins=100, hist=True, kde=True, rug=True, fit=norm, color='y', vertical=True)
```

```python
sns.distplot(df["total_bill"], bins=100, hist=True, kde=True, rug=False, fit=norm, color='r', 
             vertical=False, axlabel="Total Bill", label="Resturant")
plt.title("Explore Resturant")
plt.legend()
```

```python
plt.figure(figsize=(20,12))
sns.set()
sns.distplot(df["total_bill"], bins=100, hist=True, kde=True, rug=False, fit=norm, color='b', 
             vertical=False, axlabel="Total Bill", label="Resturant")
plt.title("Explore Resturant")
plt.legend()
```

```python
df.total_bill.sort_values()
```

```python
plt.figure(figsize=(20,12))
sns.set()
sns.distplot(df["total_bill"], bins=100, hist=True, kde=True, rug=False, fit=norm, 
             vertical=False, axlabel="Total Bill", label="Resturant", 
             hist_kws ={'color':'#DC143C', 'edgecolor':'#aaff00','linewidth':5,'linestyle':'-'})
```

```python
plt.figure(figsize=(20,12))
sns.set()
sns.distplot(df["total_bill"], bins=100, hist=True, kde=True, rug=False, fit=norm, 
             vertical=False, axlabel="Total Bill", label="Resturant", 
             hist_kws ={'color':'#DC143C', 'edgecolor':'#aaff00','linewidth':5,'linestyle':'-'},
            kde_kws = {'color':'#8e00ce', 'linewidth':5,'linestyle':'--'})
```

```python
#alpha is used for transparency
plt.figure(figsize=(20,12))
sns.set()
sns.distplot(df["total_bill"], bins=100, hist=True, kde=True, rug=False, 
             vertical=False, axlabel="Total Bill", label="Resturant", 
             hist_kws ={'color':'#DC143C', 'edgecolor':'#aaff00','linewidth':5,'linestyle':'-'},
            kde_kws = {'color':'#8e00ce', 'linewidth':5,'linestyle':'--'},
            fit=norm, 
            fit_kws = {'color':'#DC143C', 'linewidth':15,'linestyle':'--','alpha':0.20})
plt.title("Hotel", fontsize=20)
plt.xlabel("Size", fontsize=20)
plt.ylabel("Total Bill", fontsize=20)
plt.show()
```
